```python
import sys
import re
import os
import numpy as np
sys.path.insert(0, '../snow_terrain_tiles/')
sys.path.insert(0, './utils/')
from dem_utils import read_dem
from dem_utils import rasters_to_dict
from util import similarity_utils
from mpl_toolkits.axes_grid1 import make_axes_locatable
from similarity_tests.similarity_metrics import compute_similarity
import matplotlib.pyplot as plt
import rasterio
from timeit import default_timer as timer
```

```python
d = rasters_to_dict('./example_data/')
pairs = similarity_utils.create_pairs(d)
```

```python
for p in pairs.keys():
    ys = [y for y in pairs[p].keys()]
    im1 = pairs[p][ys[0]]['arr']
    im2 = pairs[p][ys[1]]['arr']
    pairs[p]['results'] = compute_similarity(im1,im2)
```

```python
# To plot the metrics
for p in pairs.keys():
    similarity_utils.plot_sim_metrics(pairs[p], p)
```

```python
similarity_utils.plot_comparison_inputs(d)
```

```python
import pandas as pd
scores = []
metrics = []
pnames = []
for p in pairs:
    pnames.append(p)
    sc = []
    for k in pairs[p]['results'].keys():
        if type(pairs[p]['results'][k]) == np.float64:
            sc.append(pairs[p]['results'][k])
            if k not in metrics:
                metrics.append(k)
    scores.append(sc)
            
#scores
df = pd.DataFrame(columns=metrics)
for result, name in zip(scores, pnames):
    df.loc[name] = result
df
```

```python
import seaborn as sns
```

```python
sns.heatmap(df, annot=True)
```

```python
no_self = df.drop(df[df.nrmse_val == 0.0].index)
sns.heatmap(no_self, annot=True)
```

```python
rdf = pd.DataFrame()
rdf['MSE Rank'] = no_self['nrmse_val'].rank(ascending=True)
rdf['SSIM Rank'] = no_self['ssim_val'].rank(ascending=False)
rdf['CW-SSIM Rank'] = no_self['cwssim_val'].rank(ascending=False)
rdf['GMSD Rank'] = no_self['gms_val'].rank(ascending=True)

rdf['Avg. Rank'] = rdf.mean(axis=1)
```

```python
sns.heatmap(rdf, annot=True)
```
