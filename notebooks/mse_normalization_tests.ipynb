```python
def mean_square_error_ordered_norms(im1, im2):
    
    mse = np.mean(np.square(im1 - im2))
    
    min_max_normalized_mse = np.sqrt(mse) / (im1.max() - im1.min())
    euclidean_normalized_mse = np.sqrt(mse) / np.sqrt(np.mean(im1**2))
    mean_normalized_mse = np.sqrt(mse) / im1.mean()
    
    return [mse, min_max_normalized_mse, euclidean_normalized_mse, mean_normalized_mse]
    
def mean_square_error_orderless_norms(im1, im2):

    im1_max = np.nanmax(im1)
    im2_max = np.nanmax(im2)
    im1_min = np.nanmin(im1)
    im2_min = np.nanmin(im2)
    both_max = max(im1_max, im2_max)
    both_min = min(im1_min, im2_min)

    mse = np.mean(np.square(im1 - im2))
    min_max_of_both_normalized_mse = np.sqrt(mse) / (both_max - both_min)
    mean_of_both_normalized_mse = np.sqrt(mse) / ((im1.mean() + im2.mean()) / 2)
    euclidean_both_normalized_mse = np.sqrt(mse) / np.sqrt(np.mean(im1*im2))
    
    return [mse, min_max_of_both_normalized_mse, euclidean_both_normalized_mse, mean_of_both_normalized_mse]
```

```python
mse_values = [round(x, 3) for x in mean_square_error_ordered_norms(sd_2012, sd_2018)]
mse_values
```

```python
# change the order, makes a difference for normalization!!
mse_values = [round(x, 3) for x in mean_square_error_ordered_norms(sd_2018, sd_2012)]
mse_values
```

```python
def mean_square_error_orderless_norms_arrs(im1, im2):

    im1_max = np.nanmax(im1)
    im2_max = np.nanmax(im2)
    im1_min = np.nanmin(im1)
    im2_min = np.nanmin(im2)
    both_max = max(im1_max, im2_max)
    both_min = min(im1_min, im2_min)

    square_error = np.square(im1 - im2)
    min_max_of_both_normalized_mse = np.sqrt(square_error) / (both_max - both_min)
    mean_of_both_normalized_mse = np.sqrt(square_error) / ((im1.mean() + im2.mean()) / 2)
    euclidean_both_normalized_mse = np.sqrt(square_error) / np.sqrt(np.abs(im1*im2))
    
    return [square_error, min_max_of_both_normalized_mse, euclidean_both_normalized_mse, mean_of_both_normalized_mse]
```

```python
orderless_mse_arrs = mean_square_error_orderless_norms_arrs(sd_2012, sd_2018)
```
