```python
import rasterio
import matplotlib.pyplot as plt
from rasterio.plot import show
plt.switch_backend('tkagg')
```

```python
src = rasterio.open('../DEMs/hv/bare_earth/orthos/hv_ortho_06_04_2017.tif')
```

```python
ortho = src.read()
```

```python
fig, ax = plt.subplots(figsize=(5,12))
show(src, ax=ax)
```

```python
xtks = ax.xaxis.get_ticklabels()
xtks1 = [str(int(str(x)[-8:-2]) - 1000) for x in xtks]

ytks = ax.yaxis.get_ticklabels()
ytks1 = [str(int(str(y)[-9:-2]) + 2000 ) for y in ytks][::-1]

print("X TICKS...", xtks1)
print("Y TICKS...", ytks1)
```

```python
fig2, ax2 = plt.subplots(figsize=(5,12))
im = show(ortho, ax = ax2)

im.set_xticklabels(xtks1)
im.set_yticklabels(ytks1)

ax2.set_title('June 4 2017 Orthomosaic')
ax2.set_ylabel('UTM E Zone 6 N')
ax2.set_xlabel('UTM N Zone 6 N')


plt.savefig('../DEMs/hv/bare_earth/figs/June_2017_ortho.png', dpi=300, bbox_inches='tight')
```
