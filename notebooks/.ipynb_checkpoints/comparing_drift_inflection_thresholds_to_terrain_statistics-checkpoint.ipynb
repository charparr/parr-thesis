```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
drifts = pd.read_csv('../drift_definition/drift_defintion_results_mean_by_study_area.to_csv')
drifts.set_index('Study Area', inplace=True)
drifts
```

```python
terrain = pd.read_csv('../DEMs/topo_roughness_stats/all_topo_stats.csv')
terrain
```

```python
terrain['Study Area'] = ['Tuolumne','Tuolumne','Tuolumne','CLPX',
                         'Tuolumne','Tuolumne','Tuolumne','Tuolumne','Happy Valley']
terrain.dropna(axis=1, inplace=True)
del terrain['Unnamed: 0']
terrain
```

```python
terrain_mean_by_study_area = terrain.copy().groupby('Study Area').mean()
terrain_mean_by_study_area
```

```python
df = terrain_mean_by_study_area.join(drifts)
df
```

```python
cols = list(df)
nunique = df.apply(pd.Series.nunique)
cols_to_drop = nunique[nunique == 1].index
df.drop(cols_to_drop, axis=1, inplace=True)
del df['Year']
```

```python
df.columns
```

```python
means = [c for c in df.columns if 'MEAN' in c]
not_shade = [c for c in means if 'shade' not in c]
mean_terrain_vals = df[not_shade]
mean_terrain_vals
```

```python
snow_props = df[['CV', 'Inflection Threshold', 'Drift Area %',
                 'Drift Volume %', 'Drift Volume %','Drift Volume-Area Difference (%)']]
snow_props
```

```python
mean_terrain_vs_snow = mean_terrain_vals.join(snow_props)
mean_terrain_vs_snow.dtypes
#sns.pairplot(snow_props, diagonal=Nonet
```
