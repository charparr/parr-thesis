```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('../subsets/agg_results/aggregate_results.csv')
df.head()
```

```python
def barchart_cwssim(df):
    fig, ax = plt.subplots(figsize=(16, 10))
    sns.barplot(x="Winters", y="cwssim", data=df, palette='gist_earth')
    ax.set_title('CW-SSIM')
    ax.set_ylim([-1, 1])
    ax.set_xticklabels(ax.get_xticklabels(),rotation=312)
#     plt.savefig('agg_results/cwssim_bars.png', dpi=300,
#                 bbox_inches='tight')
```

```python
barchart_cwssim(df)
```

```python
sns.regplot('cwssim', 'log gmsd', data=df)
```
