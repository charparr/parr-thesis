```python
import pandas as pd
import numpy as np
```

```python
ib2011 = pd.read_csv('../raw/ib2011m.csv', header=None, encoding='latin_1')
ib2011.head(11)
```

```python
ib2011.columns = ['DateTime', 'Days (AST)','1 m Wind Speed (m/s)',
                 '3 m Wind Speed (m/s)', '10 m Wind Speed (m/s)',
                 'Wind Direction (deg. True North)', '1 m Air Temp (deg. C)',
                 '3 m Air Temp (deg. C)', '10 m Air Temp (deg. C)',
                  '1 m RH (%)','3 m RH (%)', '10 m RH (%)', 'Pressure (mb)', 'Rainfall']
ib2011.drop(ib2011.index[0:10], inplace=True)
ib2011.drop('Days (AST)', axis=1, inplace=True)
ib2011.head(3)
```

```python
ib2012 = pd.read_csv('../raw/ib2012m.csv', header=None, encoding='latin_1')
ib2012.head(11)
```

```python
ib2012.drop([0,1,2], axis=1, inplace=True)
ib2012.columns = ib2011.columns
ib2012.drop(ib2012.index[0:10], inplace=True)
ib2012.head(3)
```

```python
ib2013 = pd.read_csv('../raw/ib2013m.csv', header=None)#, encoding='latin_1')
ib2013.head(12)
```

```python
ib2013.drop([1,6,7], axis=1, inplace=True)
ib2013.columns = ib2011.columns
ib2013.drop(ib2013.index[0:11], inplace=True)
ib2013.head(3)
```

```python
ib2014 = pd.read_csv('../raw/ib2014m.csv', header=None, encoding='latin_1')
ib2014.head(12)
```

```python
ib2014.drop(0, axis=1, inplace=True)
ib2014.columns = ib2011.columns
ib2014.drop(ib2014.index[0:11], inplace=True)
ib2014.head(3)
```

```python
ib2011_2014 = pd.concat([ib2011, ib2012, ib2013, ib2014])
ib2011_2014.drop(['3 m Air Temp (deg. C)', '10 m Air Temp (deg. C)',
                 '3 m RH (%)', '10 m RH (%)', 'Rainfall'], axis=1, inplace=True)
ib2011_2014.head(3)
```

```python
ib2015 = pd.read_csv('../raw/ib2015m.csv', header=None)#, encoding='latin_1')
ib2015.head(21)
```

```python
ib2015.iloc[19]
```

```python
ib2015=ib2015.copy()[[1, 9, 10, 11, 12, 3, 4, 16]]
ib2015.columns = ib2011_2014.columns
ib2015.drop(ib2015.index[0:20], inplace=True)
ib2015.head(3)
```

```python
ib2016 = pd.read_csv('../raw/ib2016m.csv', header=None)#, encoding='latin_1')
ib2016.head(21)
```

```python
ib2016=ib2016.copy()[[1, 9, 10, 11, 12, 3, 4, 16]]
ib2016.columns = ib2011_2014.columns
ib2016.drop(ib2016.index[0:20], inplace=True)
ib2016.head(3)
```

```python
ib2017 = pd.read_csv('../raw/ib2017m.csv', header=None)#, encoding='latin_1')
ib2017.head(21)
```

```python
ib2017=ib2017.copy()[[1, 9, 10, 11, 12, 3, 4, 16]]
ib2017.columns = ib2011_2014.columns
ib2017.drop(ib2017.index[0:20], inplace=True)
ib2017.head(3)
```

```python
ib2018 = pd.read_csv('../raw/ib2018m.csv', header=None)#, encoding='latin_1')
ib2018.head(21)
```

```python
ib2018=ib2018.copy()[[1, 9, 10, 11, 12, 3, 4, 16]]
ib2018.columns = ib2011_2014.columns
ib2018.drop(ib2018.index[0:20], inplace=True)
ib2018.head(3)
```

```python
ib2011_2018 = pd.concat([ib2011_2014, ib2015, ib2016, ib2017, ib2018])
```
