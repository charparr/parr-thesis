```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
imnav = pd.read_csv('../imnav_2011_2018_met_data_clean.csv')
imnav.set_index(pd.DatetimeIndex(imnav['DateTime']), inplace=True)
imnav.replace(7777, np.NaN, inplace=True)
imnav.replace(6999, np.NaN, inplace=True)
imnav['Wind Direction (deg. True North)'] = imnav['Wind Direction (deg. True North)'].round()
imnav.head()
```

```python
snow_yr_2012 = imnav['20110901':'20120420']
snow_yr_2013 = imnav['20120901':'20130420']
snow_yr_2015 = imnav['20140901':'20150420']
snow_yr_2016 = imnav['20150901':'20160420']
snow_yr_2017 = imnav['20160901':'20170420']
snow_yr_2018 = imnav['20170901':'20180420']
```

```python
def fuzzy_wx(wx_data, ws_thresh, at_thresh):
    
    wind_check = wx_data['1 m Wind Speed (m/s)'] > ws_thresh
    temp_check = wx_data['1 m Air Temp (deg. C)'] < at_thresh
    flux_possible = wind_check * temp_check
    wx_data['Flux Event Possible'] = flux_possible * wx_data['1 m Wind Speed (m/s)']

    pivot_df = wx_data.pivot(index='Wind Direction (deg. True North)',
                             columns='DateTime',
                             values='Flux Event Possible')
    
    x, y = np.meshgrid(pivot_df.columns.values, pivot_df.index.values) 
    data_values = pivot_df.values
    cumulative = np.nancumsum(data_values, axis=1) # sum over columns for each row
    
    return cumulative
```

```python
plt.figure(figsize=(12,12))
plt.imshow(fuzzy_wx(snow_yr_2012, 5, 0), aspect=12)
plt.title('Imnavait 2012 Flux Possibility Map')
plt.colorbar()
```

```python
plt.figure(figsize=(12,12))
plt.imshow(fuzzy_wx(snow_yr_2013, 5, 0), aspect=12)
plt.title('Imnavait 2013 Flux Possibility Map')
plt.colorbar()
```
